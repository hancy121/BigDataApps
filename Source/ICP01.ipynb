```
#Analysis on diagnosis column
data['diagnosis'].describe()
```

We have two unique value for column *diagnosis*

```
#counting rows having diagnosis value as 'M' and storing in a variable
count_M = len(data['diagnosis'][data['diagnosis'] =='M'])
```

```
#counting rows having diagnosis value as 'B' and storing in a variable
count_B = len(data['diagnosis'][data['diagnosis'] =='B'])
```

```
#Priniting the information about column diagnosis
print("As per dataset, we have %d records diagnosed as 'M' and %d as 'B'" %(count_M, count_B))
```

```
'''describe method tells us basic statistical chracteristics of each numerical features.
we get statistical feature of only 31 columns. Column 'diagnosis' is tyoe of non numerical so it is not listed'''
data.describe()
```

```
#we can also use below syntax to include all column having either type of object or boolean.
data.describe(include  = ['object', 'bool'])
```

```
#Average value of numerical fields for diagnosis vlaue of 'B'
data[data['diagnosis'] == 'B'].mean()
```
